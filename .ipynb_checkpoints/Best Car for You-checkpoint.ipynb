# BEST  CAR FOR YOU 
We live in a world we multiple options of products and services. As customers with access to a variety of possibilities, we get confused when having to take important decision on important things we need to buy, a car, for example.

There are plenty of variables we take into consideration for buying a car, which one is the most important? For starters, fuel shortage in Mexico has been an issue recently, long lines at the gas station make us rethink if buying our car model was the best decision we could have made. 

This analysis will help you, confused buyer, to choose the best car for your budget based on fuel performance. 



```python
# Dependencies
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# Load in File from resources
# 'movie_scores.csv'
data_file=("csv1.csv")
data_file_pd=pd.read_csv(data_file, encoding="latin-1")
```

```python
# Read and display with pandas
data_file_pd.head()
```

```python
data_file_two=("csv2.csv")
data_file_pd_two=pd.read_csv(data_file_two, encoding="latin-1")
```

```python
data_file_pd_two.head()
```

```python
data_file_three=("versionesypotencias.csv")
data_file_pd_three=pd.read_csv(data_file_three, encoding="latin-1")
```

```python
data_file_pd_three.head()
```

```python
Merge= pd.merge(data_file_pd, data_file_pd_three, on="Versión", how="left")
Merge = Merge.drop('Modelo_x',1)
Merge.head(10)
```

```python
Merge_table= pd.merge(Merge, data_file_pd_two, on="Versión", how="left")
Merge_table = Merge_table.drop('Trans._x',1)
Merge_table.head(10)
```

```python
Brandsperf=Merge_table.groupby(["Marca","Categoría"])["Suma de R. Ajust. (km/l)"].mean()
brandsperf=pd.DataFrame(Brandsperf)
brandsperf_clean = brandsperf.rename(columns={'Suma de R. Ajust. (km/l)': 'Rendimiento'})
brandsperf_clean.sort_values('Rendimiento',ascending=False)
```

```python
transperf=Merge_table.groupby("Trans._y")["Suma de R. Ajust. (km/l)"].mean()
transperf=pd.DataFrame(transperf)
transperf_clean = transperf.rename(columns={'Suma de R. Ajust. (km/l)': 'Rendimiento'})
transperf_clean.sort_values('Rendimiento',ascending=False)
```

```python
brandsperf_clean['Marca'] = brandsperf_clean.index
```

```python
#Gráfica de performance de marca 

brandsperf_clean.plot.bar(x='Marca', y='Rendimiento', rot=0)
plt.title('Rendimiento por marca')
plt.savefig("Rendimiento_Marca")
```

```python
#Gráfica de performance por transmisión 

transperf_clean['Trans.'] = transperf_clean.index
transperf_clean.plot.bar(x='Trans.', y='Rendimiento', rot=0)
plt.title('Rendimiento por tipo de transmisión')
plt.savefig("Rendimiento_Transmision")
```
